```python
def roughness(v, f, r, a):
    return 2.3372 - 1.5027 * r - 0.0048 * v - 5.9241 * f + 1.8912 * a + 27.6705 * f * f + 1.6612 * a * a + 0.0033 * r * v + 0.9716 * r * f - 0.9759 * r * a + 0.0108 * v * f - 0.0015 * v * a - 9 * f * a
v_min = 400
v_max = 500
f_min = 0.05
f_max = 0.15
r_min = 0.8 
r_max = 1.2
a_min = 0.2
a_max = 0.6

def check_v(v):
    if v >= v_min and v <= v_max:
        return True
    return False

def check_r(r):
    if r >= r_min and r <= r_max:
        return True
    return False

def check_f(f):
    if f >= f_min and f <= f_max:
        return True
    return False

def check_a(a):
    if a >= a_min and a <= a_max:
        return True
    return False
```
